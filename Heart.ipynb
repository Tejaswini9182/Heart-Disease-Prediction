```python
x=df.drop(['target'],axis=1)
y=df['target']
print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(train_x.head())
print(train_y.head())
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

#create a liner regression model
model_lr= LinearRegression()

#TRAIN THE MODEL ON THE TRAINING DATA
model_lr.fit(train_x,train_y)

#predictions on the training andtest data
y_pred_train_lr=model_lr.predict(train_x)
y_pred_test_lr=model_lr.predict(test_x)

#calculate metrics(optional)
train_mse_lr=mean_squared_error(train_y,y_pred_train_lr)
test_mse_lr=mean_squared_error(test_y,y_pred_test_lr)
train_r2_lr=r2_score(train_y,y_pred_train_lr)
test_r2_lr=r2_score(test_y,y_pred_test_lr)

```

```python
print(train_mse_lr)
```

```python
print(test_mse_lr)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

#call the model
model_lr=LogisticRegression(max_iter=1000)

#train?fit the model
model_lr.fit(train_x,train_y)

#predict from the model
y_pred_train=model_lr.predict(train_x)
y_pred_test=model_lr.predict(test_x)
```

```python
print(y_pred_train)
print(y_pred_test)
```

# K Means Clustering

```python
#k Means
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
#create Kmeans Model

k=3# Number of Clusters
model_kmeans=KMeans(n_clusters=k,random_state=42) #adding random_state

#SELECT THE COLUMNS FOR CLUSTERING
columns_for_clustering=['age','chol']
df_for_clustering=df[columns_for_clustering]

#fit the model to the data
model_kmeans.fit(df_for_clustering)

#get the cluster labels
cluster_labels=model_kmeans.cluster_centers_

# get cluster centers
cluster_centers=model_kmeans.cluster_centers_

#plot the clusters
plt.figure(figsize=(8,6))
plt.scatter(df_for_clustering['age'],df_for_clustering['chol'],c=cluster_labels,cmap='viridis',alpha=0.7)
plt.scatter(cluster_centers[:,0],cluster_centers[:,1],c='red',marker='x',s=100,label='centroids')
plt.xlabel('Age')
plt.ylabel('cholesterol')
plt.title('k-Means Clustering')
plt.legend()
plt.show()

```

```python
#to remove noise we use z_score
from scipy import stats
import numpy as np
#assuming 'df' is your dataframe and
z_scores=stats.zscore(df['age']) #z=(x-mew)/sd
abs_z_scores=np.abs(z_scores) #absolute values we need to fine and filter them.
filtered_entries=(abs_z_scores<3) #if z_score is >3 it is outlier
df=df[filtered_entries]
df
```
